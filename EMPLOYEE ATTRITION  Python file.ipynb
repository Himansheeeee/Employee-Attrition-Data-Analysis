## Hypothesis Testing :
- Hypothesis testing can help you determine if there's a statistically significant relationship between employee characteristics (like job satisfaction) and attrition.

### Formulating Hypotheses
- **Null Hypothesis (H0):** Employees with low job satisfaction do not have higher attrition rates.
- **Alternative Hypothesis (H1):** Employees with low job satisfaction have higher attrition rates.


#### T-Test
1. Prepare Data:

```python
left_employees = a[a['Attrition'] == 'Yes']['JobSatisfaction']
stayed_employees = a[a['Attrition'] == 'No']['JobSatisfaction']
```

```python
from scipy import stats

t_stat, p_value = stats.ttest_ind(left_employees, stayed_employees)
```

**Interpret Results:**

- If p-value < 0.05, reject the null hypothesis. This means there is a significant difference in job satisfaction between employees who left and those who stayed.

```python
t_stat
```

```python
p_value
```

- Given that your p-value is so small, you reject the null hypothesis. This means that there is a statistically significant relationship between job satisfaction and attrition. Employees with lower job satisfaction are more likely to leave the company.

### Hypothesis Testing for Job Role and Attrition of Employees:
**Null Hypothesis (H₀):**
- There is no association between Job Level and Attrition. In other words, an employee's job level does not affect their likelihood of leaving the company.

    - H₀: Job Level and Attrition are independent.
**Alternate Hypothesis (H₁):**
- There is an association between Job Level and Attrition. This means that an employee's job level does affect their likelihood of leaving the company.

    - H₁: Job Level and Attrition are dependent (there is a relationship between the two).

```python
left_employees = a[a['Attrition'] == 'Yes']['JobLevel']
stayed_employees = a[a['Attrition'] == 'No']['JobLevel']
```

```python
from scipy import stats

t_stat, p_value = stats.ttest_ind(left_employees, stayed_employees)
```

#### Interpret Results:

- If p-value < 0.05, reject the null hypothesis. This means there is a significant difference in Job_Role between employees who left and those who stayed.

```python
p_value
```

```python
t_stat
```

#### Results:
Since the p-value > 0.05, you fail to reject the null hypothesis. This suggests that there is no significant difference in attrition across different job roles based on this test.
